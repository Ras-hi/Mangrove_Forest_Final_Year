```python
import os
import cv2
import numpy as np
import pandas as pd

# Define directories
avicennia_dir = "/content/ndvi_images/Avicennia"
ceriops_dir = "/content/ndvi_images/Ceriops"
exoecaria_dir = "/content/ndvi_images/Exoecaria"
# Output CSV file paths
output_csv_avicennia = "/content/ndvi_images/Avicennia/avicennia_ndvi_results.csv"
output_csv_ceriops = "/content/ndvi_images/Ceriops/ceriops_ndvi_results.csv"
output_csv_exoecaria = "/content/ndvi_images/Exoecaria/exoecaria_ndvi_results.csv"
# Get image file lists
avicennia_images = sorted([f for f in os.listdir(avicennia_dir) if f.endswith(".png")])
ceriops_images = sorted([f for f in os.listdir(ceriops_dir) if f.endswith(".png")])
exoecaria_images = sorted([f for f in os.listdir(exoecaria_dir) if f.endswith(".png")])
# Lists to store NDVI data
ndvi_data_avicennia = []
ndvi_data_ceriops = []
ndvi_data_exoecaria = []
# Function to compute NDVI for an image
def compute_ndvi(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading {image_path}")
        return None

    # Convert image to float32 for NDVI calculation
    image = image.astype(np.float32) / 255.0

    # Extract RED and NIR (approximating NIR using Green channel)
    red = image[:, :, 2]  # Red channel
    nir = image[:, :, 1]  # Green channel (approximate NIR)

    # Compute NDVI
    ndvi = (nir - red) / (nir + red + 1e-8)  # Avoid division by zero

    # Return mean NDVI for the image
    return np.mean(ndvi)

# Process Avicennia images
for image_file in avicennia_images:
    image_path = os.path.join(avicennia_dir, image_file)
    mean_ndvi = compute_ndvi(image_path)
    if mean_ndvi is not None:
        ndvi_data_avicennia.append({"Image": image_file, "Mean_NDVI": mean_ndvi})

# Process Ceriops images
for image_file in ceriops_images:
    image_path = os.path.join(ceriops_dir, image_file)
    mean_ndvi = compute_ndvi(image_path)
    if mean_ndvi is not None:
        ndvi_data_ceriops.append({"Image": image_file, "Mean_NDVI": mean_ndvi})
# Process Ceriops images
for image_file in exoecaria_images:
    image_path = os.path.join(exoecaria_dir, image_file)
    mean_ndvi = compute_ndvi(image_path)
    if mean_ndvi is not None:
        ndvi_data_exoecaria.append({"Image": image_file, "Mean_NDVI": mean_ndvi})
# Convert to DataFrames and save as CSV
pd.DataFrame(ndvi_data_avicennia).to_csv(output_csv_avicennia, index=False)
pd.DataFrame(ndvi_data_ceriops).to_csv(output_csv_ceriops, index=False)
pd.DataFrame(ndvi_data_exoecaria).to_csv(output_csv_exoecaria, index=False)

print(f"✅ NDVI data saved in:\n{output_csv_avicennia}\n{output_csv_ceriops}\n{output_csv_exoecaria}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import os

# Define paths for NDVI data
avicennia_csv = "/content/ndvi_images/Avicennia/avicennia_ndvi_results.csv"
ceriops_csv = "/content/ndvi_images/Ceriops/ceriops_ndvi_results.csv"
exoecaria_csv = "/content/ndvi_images/Exoecaria/exoecaria_ndvi_results.csv"  # Corrected spelling

# Define output paths
avicennia_output_csv = "/content/ndvi_images/Avicennia/avicennia_ndvi_change_detection.csv"
ceriops_output_csv = "/content/ndvi_images/Ceriops/ceriops_ndvi_change_detection.csv"
exoecaria_output_csv = "/content/ndvi_images/Exoecaria/exoecaria_ndvi_change_detection.csv"  # Corrected spelling

# Function to apply S-CCD and detect change points
def analyze_ndvi_change(data_path, output_path, species_name):
    if not os.path.exists(data_path):
        print(f"⚠️ File not found: {data_path}")
        return

    # Load NDVI data
    data = pd.read_csv(data_path)

    # Extract year from filenames (e.g., "Avicennia_2013.png")
    data['Year'] = data['Image'].str.extract(r'(\d{4})').astype(float)  # Convert year to float

    # Sort data by year
    data.sort_values(by="Year", inplace=True)
    data.set_index("Year", inplace=True)

    # Handle missing values
    data.dropna(inplace=True)

    # Apply State-Space Model for S-CCD
    model = SARIMAX(data['Mean_NDVI'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 3))
    result = model.fit(disp=False)
    data['Predicted_NDVI'] = result.fittedvalues

    # Change Point Detection using Bayesian Probability
    change_prob = np.abs(data['Mean_NDVI'] - data['Predicted_NDVI']) / (data['Mean_NDVI'] + 1e-8)
    data['Change_Probability'] = change_prob

    # Save results to a new CSV file
    data.to_csv(output_path)
    print(f"✅ NDVI change detection results saved to: {output_path}")

    # Visualization
    plt.figure(figsize=(10, 5))
    plt.plot(data.index, data['Mean_NDVI'], label='Observed NDVI', color='green', marker='o')
    plt.plot(data.index, data['Predicted_NDVI'], label='Predicted NDVI', linestyle='dashed', color='red')
    plt.fill_between(data.index, data['Mean_NDVI'], data['Predicted_NDVI'], color='gray', alpha=0.3, label='Change Region')
    plt.title(f"{species_name} NDVI Change Detection using S-CCD")
    plt.xlabel("Year")
    plt.ylabel("NDVI")
    plt.legend()
    plt.grid(True)
    plt.show()

# Analyze NDVI changes for Avicennia
analyze_ndvi_change(avicennia_csv, avicennia_output_csv, "Avicennia")

# Analyze NDVI changes for Ceriops
analyze_ndvi_change(ceriops_csv, ceriops_output_csv, "Ceriops")

# Analyze NDVI changes for Exoecaria
analyze_ndvi_change(exoecaria_csv, exoecaria_output_csv, "Exoecaria")  # Corrected spelling
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import os

# Define paths for NDVI data
avicennia_csv = "/content/ndvi_images/Avicennia/avicennia_ndvi_results.csv"
ceriops_csv = "/content/ndvi_images/Ceriops/ceriops_ndvi_results.csv"
exoecaria_csv = "/content/ndvi_images/Exoecaria/exoecaria_ndvi_results.csv"  # Added Exoecaria

# Define output paths for separate time periods
output_files = {
    "Avicennia_Pre": "/content/ndvi_images/Avicennia/avicennia_ndvi_pre_cyclone.csv",
    "Avicennia_During": "/content/ndvi_images/Avicennia/avicennia_ndvi_during_cyclone.csv",
    "Avicennia_Post": "/content/ndvi_images/Avicennia/avicennia_ndvi_post_cyclone.csv",
    "Ceriops_Pre": "/content/ndvi_images/Ceriops/ceriops_ndvi_pre_cyclone.csv",
    "Ceriops_During": "/content/ndvi_images/Ceriops/ceriops_ndvi_during_cyclone.csv",
    "Ceriops_Post": "/content/ndvi_images/Ceriops/ceriops_ndvi_post_cyclone.csv",
    "Exoecaria_Pre": "/content/ndvi_images/Exoecaria/exoecaria_ndvi_pre_cyclone.csv",  # Added Exoecaria
    "Exoecaria_During": "/content/ndvi_images/Exoecaria/exoecaria_ndvi_during_cyclone.csv",
    "Exoecaria_Post": "/content/ndvi_images/Exoecaria/exoecaria_ndvi_post_cyclone.csv",
}

# Function to analyze NDVI change detection in different time periods
def analyze_ndvi_change(data_path, species_name):
    if not os.path.exists(data_path):
        print(f"⚠️ File not found: {data_path}")
        return

    # Load NDVI data
    data = pd.read_csv(data_path)

    # Extract year from filenames (e.g., "Avicennia_2013.png")
    data['Year'] = data['Image'].str.extract(r'(\d{4})').astype(int)

    # Convert Year to DatetimeIndex
    data['Year'] = pd.to_datetime(data['Year'], format="%Y")
    data.set_index("Year", inplace=True)

    # Define time frames
    time_frames = {
        "Pre": ("2013", "2018"),
        "During": ("2019", "2020"),
        "Post": ("2021", "2023"),
    }

    for period, (start_year, end_year) in time_frames.items():
        period_data = data.loc[(data.index.year >= int(start_year)) & (data.index.year <= int(end_year))].copy()

        if period_data.empty:
            print(f"⚠️ No data available for {species_name} in {period}. Skipping.")
            continue

        # Apply State-Space Model for S-CCD
        model = SARIMAX(period_data['Mean_NDVI'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 3))
        result = model.fit(disp=False)
        period_data['Predicted_NDVI'] = result.fittedvalues

        # Change Point Detection using Bayesian Probability
        change_prob = np.abs(period_data['Mean_NDVI'] - period_data['Predicted_NDVI']) / (period_data['Mean_NDVI'] + 1e-8)
        period_data['Change_Probability'] = change_prob

        # Save results to CSV
        output_path = output_files[f"{species_name}_{period}"]
        period_data.to_csv(output_path)
        print(f"✅ NDVI change detection results saved for {species_name} in {period}: {output_path}")

        # Visualization
        plt.figure(figsize=(10, 5))
        plt.plot(period_data.index.year, period_data['Mean_NDVI'], label='Observed NDVI', color='green', marker='o')
        plt.plot(period_data.index.year, period_data['Predicted_NDVI'], label='Predicted NDVI', linestyle='dashed', color='red')
        plt.fill_between(period_data.index.year, period_data['Mean_NDVI'], period_data['Predicted_NDVI'], color='gray', alpha=0.3, label='Change Region')
        plt.title(f"{species_name} NDVI Change Detection ({period}-Cyclone)")
        plt.xlabel("Year")
        plt.ylabel("NDVI")
        plt.legend()
        plt.grid(True)
        plt.show()

# Analyze NDVI changes for Avicennia
analyze_ndvi_change(avicennia_csv, "Avicennia")

# Analyze NDVI changes for Ceriops
analyze_ndvi_change(ceriops_csv, "Ceriops")

# Analyze NDVI changes for Exoecaria
analyze_ndvi_change(exoecaria_csv, "Exoecaria")  # Added Exoecaria
```
